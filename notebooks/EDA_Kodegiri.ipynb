```python
import pandas as pd
import re
```

# Deals

```python
df_deals = pd.read_excel('deals.xlsx')
df_deals.head()
```

```python
df_deals.info()
```

```python
df_deals = df_deals.iloc[:,1:]
```

```python
df_deals['Amount'] = df_deals['Amount'].astype(str)
df_deals['Amount'] = df_deals['Amount'].apply(lambda x: re.sub(r'\D', '', x))
df_deals['Amount'] = df_deals['Amount'].replace('', '0') # Replace empty strings with '0'
df_deals['Amount'] = df_deals['Amount'].astype(int)
```

```python
df_deals['Expected Closed Date'] = pd.to_datetime(df_deals['Expected Closed Date'])
```

```python
df_deals.info()
```

```python
df_deals = df_deals.dropna(subset=['Deal Name'])
```

# Activities

```python
df_activities = pd.read_excel('Activities.xlsx')
df_activities.head()
```

```python
df_activities.info()
```

```python
df_activities = df_activities.iloc[:,1:]
```

```python
df_activities['Date'] = pd.to_datetime(df_activities['Date'])
```

```python
df_activities.info()
```

# Contacts

```python
df_contacts = pd.read_excel('contacts.xlsx')
df_contacts.head()
```

```python
df_contacts.drop(columns=['No'], inplace=True)
df_contacts = df_contacts.iloc[1: , :]
```

```python
df_contacts.head()
```

```python
df_contacts.info()
```

```python
df_contacts = df_contacts.iloc[:147, :]
```

```python
df_contacts = df_contacts.reset_index(drop=True)
```

```python
# Menampilkan baris yang memiliki nilai kosong
missing_data = df_contacts[df_contacts['First Name'].isnull()]

# Menampilkan DataFrame dengan baris yang memiliki nilai kosong
print(missing_data)
```

```python
df_contacts = df_contacts.dropna(subset=['First Name'])
```

```python
df_contacts.info()
```

# Merge 1 (deals activities)

```python
merged_df = pd.merge(df_deals, df_activities, left_on='Deal Name', right_on='Deals', how='inner')
merged_df
```

```python
merged_df.info()
```

```python
merged_df = merged_df.drop(['Owner_y', 'Contacts_y', 'Companies_y', 'Deals'], axis=1)
```

# Merge 2 (merge1 Contacts)

```python
merged_df2 = pd.merge(merged_df, df_contacts, left_on='Contacts_x', right_on='Full Name', how='left')
merged_df2
```

```python
merged_df2.info()
```

```python
merged_df2 = merged_df2.iloc[:,:18]
merged_df2 = merged_df2.drop(['First Name', 'Last Name', 'Email', 'Phone Number'], axis=1)
```

```python
merged_df2.info()
```

```python
merged_df2.describe()
```
